```python
x = np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x, np.sin(x)); # 상태 기반 인터페이스
```

axis 축

axes 축의 복수

```python
ax=plt.axes()
ax.plot(x, np.sin(x)) # 객체 지향 인터페이스
```

```python
plt.plot(x, np.sin(x), color='red', label='sin(x)')
plt.plot(x, np.cos(x), color='g', linestyle='--', label='cos(x)')
plt.title('Sin & Cos curves')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim((-1.2,1.2))
plt.tight_layout()
```
